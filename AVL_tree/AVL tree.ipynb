```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
        self.height_of_left=0
        self.height_of_right=0
class AVL:
    def __init__(self):
        self.root=None
        self.liste=[-1,0,1]
    def search(self,data):
        tmp=self.root
        while tmp != None:
            if tmp.data==data:
                return True
            elif tmp.data>data:
                tmp=tmp.left
            else:
                tmp=tmp.right
        return False
    def toplam(self,current):
        top=current.height_of_left-current.height_of_right
        return top
    def insertion(self,current,node_data):
        if current==None:
            return node_data
        if current.data>node_data.data:
            current.left=self.insertion(current.left,node_data)
            current.height_of_left=self.height(current.left)
        else:
            current.right=self.insertion(current.right,node_data)
            current.height_of_right=self.height(current.right)
        top=self.toplam(current)
        if top in self.liste:
            return current
        else:
            top1=current.height_of_left-current.height_of_right
            if current.left!=None:
                top2=current.left.height_of_left-current.left.height_of_right
            if current.right!=None:
                top3=current.right.height_of_left-current.right.height_of_right
            if top1==2 and top2==1:
                return self.RR_Rotation(current)
            elif top1==-2 and top3==-1:
                return self.LR_Rotation(current)
            elif top1==2 and top2==-1:
                return self.LRR_Rotation(current)
            elif top1==-2 and top3==1:
                return self.RLR_Rotation(current)
    def add(self,data):
        node_data=Node(data)
        if self.root==None:
            self.root=node_data
        else:
            self.insertion(self.root,node_data)       
    def height(self,current):
        return 1+max(current.height_of_right,current.height_of_left)
    def RR_Rotation(self,node_data):
        b=node_data.left
        tmp=b.right
        b.right=node_data
        node_data.left=tmp
        return b
    def LR_Rotation(self,node_data):
        b=node_data.right
        tmp=b.left
        b.left=node_data
        node_data.right=tmp
        return b
    def LRR_Rotation(self,node_data):
        node_data.left=self.LR_Rotation(node_data.left)
        return self.RR_Rotation(node_data)
    def RLR_Rotation(self,node_data):
        node_data.right=self.LR_Rotation(node_data.right)
        return self.LR_Rotation(node_data)
    def to_print(self):
        if self.root==None:
            print("ağacımız boş")
        else:
            current=self.root
            self.preorder(current)
    def preorder(self,current):
        if current==None:
            return 
        print(current.data)
        self.preorder(current.left)
        self.preorder(current.right)
    def postorder(self,current):
        if current==None:
            return 
        self.preorder(current.left)
        self.preorder(current.right)
        print(current.data)
    def inorder(self,current):
        if current==None:
            return 
        self.inorder(current.left)
        print(current.data)
        self.inorder(current.right)
    def loop(self,liste):
        for i in range(len(liste)):
            self.add(liste[i])
avl=AVL()
lst=[120,100,140,80,60,90]
avl.loop(lst)
avl.to_print()
#ماصار الكود كلو صح بعدين برحعلو 
```
